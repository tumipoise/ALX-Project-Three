## Investigation Overview

#### In this investigation, I studied loan features that could aid in determining the borrower's rate and loan status. In terms of loans, my primary focus was on the prosper risk score, the borrower's monthly income, and the borrower's employment status.

## Dataset Overview

#### The 113,937 loans in the data set used for this study have an average of 81 variables, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. However, only 5 features were used because they were relevant to my analysis, and about 30,000 missing entries were removed because the data was collected before July 2009 and did not include prosper score data ( a very important variable for my analysis) The prosper score is a custom risk score built using historical Prosper data that is used to scale loans based on how risky or less risky they are. The risk score ranges from 1 to 10, with 10 being the best. Loans originated are eligible.

## Distribution of Loan Status

#### The distribution of Loan status shows the loan status ranges from 0 to 20000 and the majority of loans are completed with a count of about 19500, charged off loans are 5000 while defaulted loans are about 750 which has the least count.

## Distribution of Prosper Score

#### The Prosper risk score distribution reveals that its values range from 1 to 11 on both the bar chart and histogram respectively, with the majority (the mode) of loans having a risk score of 8.0, which, according to the scale of the prosper score, appears to be risky. However, we would notice high counts at prosper scores of 6.0, which indicate less risky loans.

## Distribution of Employment Status

#### The employment status distribution shows that the borrowers that are employed have the highest count of 16000, followed by borrowers that are full-time with 6000 count, compared to the borrowers that are unemployed, having count of about 500.

## Distribution of Borrower Rate

#### The borrower rate distribution shows that there is a spike in the amount of loans at BorrowerRate of about 0.32 and the frequency quickly trails off  until the next spike.

## Relationship between Borrower Rate and Prosper Risk Score

#### Higher Prosper Score(Less Risky Loans) are attributed to Lower Borrower Rate (Loans with Less Returns) and vice versa as we can see that a risk score of 1(highly risky) on average have corresponds to a 31% borrower rate while a risk score of 10( less risky) corresponds to a borrower rate of 8%.

## Relationship between Employment Status and Borrower Rate

#### Looking specifically at employed and unemployed borrowers, employed borrowers have a 22% borrower rate compared to unemployed borrowers' 25% borrower rate.

## Relationship between loan status and prosper risk scores

#### The above boxplot shows that the defaulted loans have a risk score of 6, the completed loan have a risk score of 7 and the chargedoff loan have a risk score of 5.

## Loan status, Prosper Risk Score , Employment Status

#### With regard to defaulted, completed, and charged-off loans, we can state with certainty that the risk score for unemployed people's loans always lags behind that of all employed people. This indicates that the most risky current, completed, and charged-off loans are collected by unemployed people, while the most risky past-due loans were collected by employed people, as the risk score for employed people shows.
